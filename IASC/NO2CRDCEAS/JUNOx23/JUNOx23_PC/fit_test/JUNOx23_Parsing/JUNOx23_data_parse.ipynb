```python
import datetime as dt
import pytz
import csv
import pandas as pd
```

```python
######
## FROM 10am - 6am UTC and >6am to <10am UTC
#

filename = 'MC230626pm.txt'
dayutc = dt.datetime(2023,6,26, tzinfo=pytz.utc)

with open(filename) as f:
    lines = f.readlines()

data = []

for line in lines:
    a = line.strip('\n')
    data.append(a.split(' '))

col1 = []
col2 = []
col3 = []
for ele in data:
    timestamp = dt.datetime.strptime(ele[0],'%Y/%m/%d-%H:%M:%S')
    minutes = (timestamp.astimezone(pytz.timezone('Europe/Berlin'))-dayutc).total_seconds()/60
    col1.append(minutes)
    col2.append(float(ele[1]))
    col3.append(int(ele[-1]))
    


df1 = pd.DataFrame(list(zip(*[col1, col2, col3]))).add_prefix('Col')
df1.columns = ['Time', 'NO2', 'Flag']
df_inv =df1.fillna(int(-999))
df_inv.NO2[df_inv.NO2==0]=-999
#df_inv.to_csv('2023-06-24.COR_CRDS.am.dat', index=False,header=True)
```

```python
df_filter = df1
tempi = None
for i in df_filter.index:
    if df_filter.NO2[i] == 0:
        df_filter.NO2[i] = -999
        if df_filter.NO2[i+1] != 0:
            tempi = i
    try:
        if i <= tempi+5:
            df_filter.NO2[i] = -999
    except:
        continue

df_filter.to_csv('2023-06-26.COR_CRDS.pm.dat', index=False,header=True)
```
