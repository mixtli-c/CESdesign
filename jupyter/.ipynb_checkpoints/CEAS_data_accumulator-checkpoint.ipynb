```python
import numpy as np
import datetime as dt
import scipy.signal as scs
import pandas as pd
import glob
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import sys
sys.path.append('D:\\gitCCA\\CESdesign\\py')
import CESfunctions_dev as cf
```

```python
##### RAMA Accumulator

folder="D:\\CCA\\ICACC-BBCEAS\\RAMA\\"
filelist=glob.glob(folder+"*.csv")
firstdate=filelist[0][len(folder)+4:len(folder)+12]
lastdate=filelist[-1][len(folder)+4:len(folder)+12]
Accum_filename=folder+"RAMA_"+firstdate+"_"+lastdate+".txt"
dateR=[]
ppbR=[]

for file in filelist:
    print("Now adding file: ",file)
    dateRn,ppbRn = cf.RAMA_read(file)
    dateR.extend(dateRn)
    ppbR.extend(ppbRn)
    print("Size of lists: %i, %i " %(len(dateR),len(ppbR)))

dateRtxt = []
for ele in dateR:
    dateRtxt.append(ele.strftime("%Y/%m/%d-%H:%M:%S"))

newdata=[dateRtxt,ppbR] 
with open(Accum_filename,'w') as f:
    for x in zip(*newdata):
        f.write("{0} {1}\n".format(*x))     
```

```python
##### Mfile accumulator

dirlist =[]
rootdir = 'D:\\CCA\\ICACC-BBCEAS\\dataset1\\'
Mfile_accum = "Mfile_accum_r_dataset.txt"

for path in glob.glob(rootdir+'*\\'):
    dirlist.append(path)

dateM=[]
ppb1M=[]
ppb2M=[]
intM=[]

for folder in dirlist[:-1]:
    print("Now working on folder: ",folder)
    Mrfile=glob.glob(folder+"*r.txt")
    dateMn,ppb1Mn,ppb2Mn,intMn = cf.Mfile_read(Mrfile[0])
    dateM.extend(dateMn)
    ppb1M.extend(ppb1Mn)
    ppb2M.extend(ppb2Mn)
    intM.extend(intMn)
    print("Size of lists is: %i,%i,%i,%i" %(len(dateM),len(ppb1M),len(ppb2M),len(intM)))

date=[]
for ele in dateM:
    date.append(ele.strftime("%Y/%m/%d-%H:%M:%S"))
    
    
newdata=[date,ppb1M,ppb2M,intM]
    
with open(rootdir+Mfile_accum,'w') as f:
    for x in zip(*newdata):
        f.write("{0} {1} {2} {3}\n".format(*x)) 
```
