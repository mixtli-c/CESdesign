```python
import numpy as np
import datetime as dt
import glob
import pandas as pd
import matplotlib.pyplot as plt
import scipy.signal as scs
from matplotlib.dates import DateFormatter
import sys
sys.path.append('/home/mixtli/localJupyter/py/')
import CESfunctions_dev as cf
```

```python
ramafolder = '/home/mixtli/localJupyter/BBCEAS_data/RAMA/'
mfolder = '/home/mixtli/localJupyter/BBCEAS_data/'

```

```python
ramafile = 'CCA_20220315_M.csv'
mfile = '20220315/M2203161014.txt'
dateM,ppb1M,ppb2M,intM = cf.Mfile_read(mfolder+mfile)
dateR,ppbR = cf.RAMA_read(ramafolder+ramafile)
```

```python
dateR2,ppbR2 = cf.RAMA_read(ramafolder+'CCA_20220316_M.csv')
```

```python
f = plt.figure(figsize=(14,14))
ramalow=700
ramahigh=900

ax1 = f.add_subplot(111)
ax1.plot(dateM,ppb1M,'--k',alpha=.25, label="BBCEAS")
ax1.plot(dateM,scs.savgol_filter(ppb1M,21,3),'-k', label="BBCEAS (smoothed)")
#ax1.plot(dateR[ramalow:ramahigh],ppbR[ramalow:ramahigh],'-r', label='RAMA')
ax1.plot(dateR,ppbR,'-r', label='RAMA')
ax1.plot(dateR2[0:900],ppbR2[0:900],'-r', label='RAMA')
ax1.xaxis.set_major_formatter(DateFormatter("%H:%M:%S"))
ax1.set_xlabel('Time',size=14)
ax1.set_ylabel('NO$_2$ ppb',size=14)
ax1.set_title("NO$_2$ measurements, March 15-16, 2022",size=16)
ax1.legend(fontsize=12, loc='upper left')
ax1.set_ylim(0,65)

plt.savefig("March15_16.png")
```

```python
bg=np.loadtxt('/home/mixtli/localJupyter/BBCEAS_data/20220310/Ib2203101539.txt')
bg2=np.loadtxt('/home/mixtli/localJupyter/BBCEAS_data/20220311/Ib2203111214.txt')
bg3=np.loadtxt('/home/mixtli/localJupyter/BBCEAS_data/20220315/Ib2203151258.txt')
bg4=np.loadtxt('/home/mixtli/localJupyter/BBCEAS_data/20220316/Ib2203161314.txt')
bg5=np.loadtxt('/home/mixtli/localJupyter/BBCEAS_data/20220316/Ib2203161507.txt')
bg6=np.loadtxt('/home/mixtli/localJupyter/BBCEAS_data/20220317/Ib2203171326.txt')
f = plt.figure(figsize=(14,14))


ax1 = f.add_subplot(211)
ax1.plot(bg[:,0],np.average(bg[:,1:],axis=1),'-k', label='Average March 10')
ax1.plot(bg2[:,0],np.average(bg2[:,1:],axis=1),'-r', label='Average March 11')
ax1.plot(bg3[:,0],np.average(bg3[:,1:],axis=1),'-b', label='Average March 15')
ax1.plot(bg4[:,0],np.average(bg4[:,1:],axis=1),'-g', label='Average March 16')
ax1.plot(bg5[:,0],np.average(bg5[:,1:],axis=1),'-y', label='Average March 16')
ax1.plot(bg6[:,0],np.average(bg6[:,1:],axis=1),'-c', label='Average March 17')
ax1.set_xlabel('Wavelength',size=14)
ax1.set_ylabel('counts',size=14)
ax1.set_title("BBCEAS Overnight Background Test : March 10/15/16",size=16)
ax1.set_xlim(440,465)
#ax1.set_ylim(650000,750000)
ax1.legend(fontsize=12, loc='upper left')

ax2 = f.add_subplot(212)
ax2.plot(bg[:,0],np.std(bg[:,1:],axis=1),'-k', label='StDev March 10')
ax2.plot(bg2[:,0],np.std(bg2[:,1:],axis=1),'-r', label='StDev March 11')
ax2.plot(bg3[:,0],np.std(bg3[:,1:],axis=1),'-b', label='StDev March 15')
ax2.plot(bg4[:,0],np.std(bg4[:,1:],axis=1),'-g', label='StDev March 16')
ax2.plot(bg5[:,0],np.std(bg5[:,1:],axis=1),'-y', label='StDev March 16')
ax2.plot(bg6[:,0],np.std(bg6[:,1:],axis=1),'-c', label='StDev March 17')
ax2.set_xlabel('Wavelength',size=14)
ax2.set_ylabel('counts',size=14)
ax2.set_title("BBCEAS Overnight Background Test : March 10/15/16",size=16)
ax2.set_xlim(440,465)

ax2.legend(fontsize=12, loc='upper left')
plt.savefig("March10_to_16-BG.png")
```
