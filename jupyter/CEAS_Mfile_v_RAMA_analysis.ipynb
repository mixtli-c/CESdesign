```python
ramafolder = 'D:\\CCA\\Avantes\\BBCEAS_Data\\RAMA\\'
mfolder = 'D:\\CCA\\Avantes\\BBCEAS_Data\\'

ramafile = 'CCA_20220202_M.csv'
mfile = '20220202\\M2202021416.txt'


dateM,ppb1M,ppb2M,intM = cf.Mfile_read(mfolder+mfile)
dateR,ppbR = cf.RAMA_read(ramafolder+ramafile)


f = plt.figure(figsize=(14,14))

ramalow=650
ramahigh=900

ax1 = f.add_subplot(111)
ax1.plot(dateM,ppb1M,'--k',alpha=.25, label="BBCEAS")
ax1.plot(dateM,scs.savgol_filter(ppb1M,21,3),'-k', label="BBCEAS (smoothed)")
ax1.plot(dateR[ramalow:ramahigh],ppbR[ramalow:ramahigh],'-r', label='RAMA')
ax1.set_ylim(-5,25)



ax1.set_xlabel('Time',size=14)
ax1.set_ylabel('NO$_2$ ppb',size=14)
ax1.set_title("Measurements, Feb-02-2022",size=16)
ax1.legend(fontsize=12, loc='upper left')

ax1.xaxis.set_major_formatter(DateFormatter("%H:%M:%S"))

#plt.savefig("Feb02-1.png")

```
